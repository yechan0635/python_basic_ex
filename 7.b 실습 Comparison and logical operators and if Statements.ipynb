## 3. 유저한테 점수를 입력받아서,

- 점수가 0보다 작거나 100보다 크면, "잘못 입력했습니다."를 출력.
- 점수가 80 이상이면, "입력하신 점수 ??는  A 학점입니다." 를 출력
- 점수가 80 점 미만이면, "입력하신 점수 ??는 B 학점입니다."를 출력

```python
score = int(input("점수를 입력하세요."))
```

```python
if score < 0 or score > 100:
  print("잘못 입력했습니다.")
elif score >= 80:
  print(f"입력하신 점수 {score}는 A학점입니다.")
elif score < 80:
  print(f"입력하신 점수 {score}는 B학점입니다.")
```

## 4. 바람의 속도를 입력받아서, 해당 바람의 속도가 다음과 같을 경우, 해당되는 문자열을 출력하세요.

### 만약 유저가 1보다 작은 값을 입력하면, "1보다 큰 수를 입력하세요"를 출력합니다.

image.png

```python
wind_spd = int(input('풍속을 입력하세요'))
```

```python
if wind_spd < 1:
  print("1이상의 수를 입력하세요.")
elif wind_spd <= 3:
  print(f'wind speed {wind_spd} : light air')
elif wind_spd <= 27:
  print(f'wind speed {wind_spd} : Breeze')
elif wind_spd <= 47:
  print(f'wind speed {wind_spd} : Gale')
elif wind_spd <= 63:
  print(f'wind speed {wind_spd} : Storm')
else:
  print(f'wind speed {wind_spd} : hurricane')
```

## 5. 회원가입을 하려 합니다.

- 유저한테 이메일과 비밀번호를 입력받습니다.
- 유저한테 이메일을 입력받습니다. 이때 이메일주소 형식(@)이 잘못 되었으면, 다시 입력하세요 를 출력합니다.
- 이메일 주소가 잘 입력되었으면, 이제 비밀번호를 두번 입력받습니다.
- 만약 첫번째 입력한 비밀번호와, 두번째 입력한 비밀번호가 다르면, 비밀번호가 일치하지 않습니다 를 출력합니다.

```python
email = "abc.naver.com"
print('@' in email)
print('@' not in email)
```

```python
user_input = input("이메일 주소 입력>")
email = user_input
# email = "abc@naver.com"

if '@' not in email :
  print("이메일을 다시 입력하세요.")
else:
  pw1 = input("패스워드 입력 >")
  pw2 = input("패스워드 확인 입력 >")

  # 만약 첫번째 입력한 비밀번호와, 두번째 입력한 비밀번호가 다르면,
  if pw1 != pw2:
    print("비밀번호가 일치하지 않습니다.")
```

## 비밀번호의 길이는 4자리 이상 8자리 이하로만 받습니다.
- 비밀번호의 길이는 4자리 이상 8자리 이하로만 받습니다.
- 그렇지 않을 경우는 "비밀번호 길이는 4자리 이상 8자리 이하로 입력하세요" 라고 출력.

```python
pw1 = input("패스워드 입력 >")
# pw1 = "123445678"
if len(pw1) < 4 or len(pw1) > 8:
  print("비밀번호 길이는 4자리 이상 8자리 이하로 입력하세요")
```

## 5번 문제 통합

```python
user_input = input("이메일 주소 입력>")
email = user_input
# "abc@naver.com"

if '@' not in email :
  print("이메일을 다시 입력하세요.")

else:
  pw1 = input("패스워드 입력 >")
  pw2 = input("패스워드 확인 입력 >")
  if len(pw1) < 4 or len(pw1) > 8:
    print("비밀번호 길이는 4자리 이상 8자리 이하로 입력하세요")
  elif pw1 != pw2:
    print("비밀번호가 일치하지 않습니다.")
  else:
    print("비밀번호가 일치합니다.")
```
