# <font color='blue'> 9. 유저가 점수를 입력하면, 해당 학점을 출력하는 프로그램입니다.  
# <font color='blue'> 단, 유저가 -1 을 입력하면, 프로그램을 종료합니다. -1 을 입력하기 전에는 계속 입력을 받습니다.


- 사용자가 어떤 값을 입력함, 그 값에 따라 A, B, C, F 등급을 표시함
- 숫자 <= 73 : F
- 74 <= 숫자 < 84 : C
- 84 <= 숫자 < 94 : B
- 94 <= 숫자 < 100 : A

```python
#  -1을 입력하기 전에는 계속 입력을 받습니다.
```

```python
while True:
    s = input("점수 입력 (-1 입력 시 종료): ").strip()

    # 정수만 허용
    try:
        score = int(s)
    except ValueError:
        print("정수로 입력해주세요. 예) 73, 84, 95")
        continue

    # 종료 조건
    if score == -1:
        print("프로그램을 종료합니다.")
        break

    # 학점 계산
    if score <= 73:
        grade = "F"
    elif 74 <= score < 84:
        grade = "C"
    elif 84 <= score < 94:
        grade = "B"
    elif 94 <= score < 100:
        grade = "A"
    else:
        # 규칙 밖의 값 처리 (예: 100 이상, 그 외 음수 등)
        print("정의되지 않은 점수 구간입니다. (규칙: A는 94 이상 100 미만)")
        continue

    print(f"점수: {score} → 학점: {grade}")
```

# <font color='blue'> 12. 해당 sms 문장에 스팸으로 등록한 단어가 있는지 검사하세요.
## 스팸 단어 리스트 = [ '대출', '금리', '도박', '잭팟']
## 검사한 후, 스팸 단어가 있으면, "스팸"을 출력하고, 없으면 "스팸 아님"을 출력하세요.

## 문장 예 ) "최저 금리로 당일 송금, 대출 100만~3000만원 가능"

```python
spam_words = [ '대출', '금리', '도박', '잭팟']

sms = "아 졸리다, 집가고 싶다"
```

```python
# sms = "아 졸리다, 집가고 싶다"
sms = "최저 금리로 당일 송금, 대출 100만~3000만원 가능"
spam_cnt = 0
for word in spam_words:
  if word in sms:
    spam_cnt = spam_cnt + 1
    break
  else:
    pass
if spam_cnt > 0:
    print("스팸")
    print("이메일 보내기")
else:
    print("스팸 아님")
```
